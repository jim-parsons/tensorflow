```python
import glob
import pickle

import matplotlib.pyplot as plt
```

```python
history_file = './history/train_demo/captcha_adam_binary_crossentropy_bs_100_epochs_10.h5'
with open(history_file, 'rb') as f:
    history = pickle.load(f)
```

```python
fig = plt.figure()
plt.subplot(2, 1, 1)
plt.plot(history['acc'])
plt.plot(history['val_acc'])
plt.title('Model Acc')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='lower right')

plt.subplot(2,1,2)
plt.plot(history['loss'])
plt.plot(history['val_loss'])
plt.title('Model Loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper right')
plt.tight_layout()

plt.show()
```
