```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

df1 = pd.read_csv('data1.csv', names=['square', 'bedroom', 'price'])
df1.head()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.set_xlabel('square')
ax.set_ylabel('bedroom')
ax.set_zlabel('price')

ax.scatter3D(df1['square'], df1['bedroom'], df1['price'], c=df1['price'], cmap='Greens')
```

```python
# 数据归一化
```

```python
def normalize_feature(df):
    return df.apply(lambda col: (col - col.mean()) / col.std())
```

```python
df = normalize_feature(df1)
df
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.set_xlabel('square')
ax.set_ylabel('bedroom')
ax.set_zlabel('price')

ax.scatter3D(df['square'], df['bedroom'], df['price'], c=df['price'], cmap='Greens')
```

```python
import numpy as np
ones = pd.DataFrame({'ones': np.ones(len(df))})
```

```python
df = pd.concat([ones, df], axis=1)
df.head()
```
